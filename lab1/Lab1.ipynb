```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(style="ticks")
```

```python
from sklearn.datasets import load_iris
```

```python
raw_data = load_iris()
features = pd.DataFrame(data=raw_data['data'],columns=raw_data['feature_names'])
data = features
data['target']=raw_data['target']
data['class']=data['target'].map(lambda ind: raw_data['target_names'][ind])
data.head()
```

```python
data.describe()
```

```python
data.shape
```

Датасет включает в себя 5 атрибутов:
Sepal length (cm)
Sepal width (cm)
Petal length (cm)
Petal width (cm)
Target

```python
data.hist()
```

```python
corr = data.corr()
```

```python
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

Наибольшая корреляция наблюдается между парами параметров petal width (cm) - petal length (cm) и petal width (cm) - target

```python
sns.jointplot(x='petal length (cm)', y='petal width (cm)', data=data)
```

```python
sns.violinplot(x=data['target'])
```

```python
for i in data.target.unique():
    sns.distplot(data['sepal length (cm)'][data.target==i],
                 kde=1,label='{}'.format(i))

plt.legend()
```
